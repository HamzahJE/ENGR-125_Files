Import traffic data from http://drnam.org/pythonfiles/traffic.csv 

The data headings are “Date	Time	Humidity Temperature	Windspeed	Car	Truck	Winddirec	NOx	PM2.5”





```
import pandas as pd
address = 'http://drnam.org/pythonfiles/traffic.csv'

dataset = pd.read_csv(address)

dataset.columns = ['Date',	'Time',	'Humidity', 'Temperature', 'Windspeed',	'Car',	'Truck',	'Winddirec',	'NOx',	'PM2_5']

```


**1)	Create a statistical report for “Humidity Temperature	Windspeed	Car	Truck	winddirec	NOx	PM2.5” The report should include Minimum, Maximum, Mean, 25 Percentile, 50 Percentile, and 75 Percentile**

```
import numpy as np
hum=dataset.Humidity
temp=dataset.Temperature
ws=dataset.Windspeed
carr=dataset.Car
trucks=dataset.Truck
wd=dataset.Winddirec
nox=dataset.NOx
pm25=dataset.PM2_5
time=dataset.Time

print("Minimum Humidity:    ",hum.min())
print("Maximum Humidity:    ",hum.max())
print("Mean Humidity:    ",hum.mean())
print("25th percentile humidity:    ", np.percentile(hum, 25))
print("50th percentile humidity:    ", np.percentile(hum, 50))
print("75th percentile humidity:    ", np.percentile(hum, 75))
```

```
print("Minimum Temp:    ",temp.min())
print("Maximum Temp:    ",temp.max())
print("Mean Temp:    ",temp.mean())
print("25th percentile Temp:    ", np.percentile(temp, 25))
print("50th percentile Temp:    ", np.percentile(temp, 50))
print("75th percentile Temp:    ", np.percentile(temp, 75))
```

```
print("Minimum Windspeed:    ",ws.min())
print("Maximum Windspeed:    ",ws.max())
print("Mean Windspeed:    ",ws.mean())
print("25th percentile Windspeed:    ", np.percentile(ws, 25))
print("50th percentile Windspeed:    ", np.percentile(ws, 50))
print("75th percentile Windspeed:    ", np.percentile(ws, 75))
```

```
print("Minimum Car :    ",carr.min())
print("Maximum Car:    ",carr.max())
print("Mean Car:    ",carr.mean())
print("25th percentile Car:    ", np.percentile(carr, 25))
print("50th percentile Car:    ", np.percentile(carr, 50))
print("75th percentile Car:    ", np.percentile(carr, 75))
```

```
print("Minimum Truck :    ",trucks.min())
print("Maximum Truck:    ",trucks.max())
print("Mean Truck:    ",trucks.mean())
print("25th percentile Truck:    ", np.percentile(trucks, 25))
print("50th percentile Truck:    ", np.percentile(trucks, 50))
print("75th percentile Truck:    ", np.percentile(trucks, 75))
```

```
print("Minimum Winddirection :    ",wd.min())
print("Maximum Winddirection:    ",wd.max())
print("Mean Winddirection:    ",wd.mean())
print("25th percentile Winddirection:    ", np.percentile(wd, 25))
print("50th percentile Winddirection:    ", np.percentile(wd, 50))
print("75th percentile Winddirection:    ", np.percentile(wd, 75))
```

```
print("Minimum NOx :    ",nox.min())
print("Maximum NOx:    ",nox.max())
print("Mean NOx:    ",nox.mean())
print("25th percentile NOx:    ", np.percentile(nox, 25))
print("50th percentile NOx:    ", np.percentile(nox, 50))
print("75th percentile NOx:    ", np.percentile(nox, 75))
```

```
print("Minimum PM2.5 :    ",pm25.min())
print("Maximum PM2.5:    ",pm25.max())
print("Mean PM2.5:    ",pm25.mean())
print("25th percentile PM2.5:    ", np.percentile(pm25, 25))
print("50th percentile PM2.5:    ", np.percentile(pm25, 50))
print("75th percentile PM2.5:    ", np.percentile(pm25, 75))
```
