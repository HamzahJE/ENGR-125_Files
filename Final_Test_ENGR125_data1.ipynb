```
import numpy as np
import pandas as pd
import seaborn as sns
```

*Download* dataset: fish.csv

```
url ='http://drnam.org/pythonfiles/Fish.csv'
raw_dataset = pd.read_csv(url)
dataset = raw_dataset.copy()
dataset.tail()
```

Step 3: Map them numerically

```
dataset['Species'] = dataset['Species'].map({'Bream': 1, 'Parkki': 2, 'Perch': 3, 'Pike': 4, 'Roach': 5, 'Smelt': 6, 'Whitefish': 7})
dataset.tail()
```

Step 4: Pairplots

```
sns.pairplot(dataset[['Species','Weight', 'Length1', 'Length2', 'Length3','Height','Width']], diag_kind='kde')
```
